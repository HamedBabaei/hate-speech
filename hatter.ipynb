# Loading Data

```python
#loading dataset
 
import utils

path = "drive/MyDrive/HateSpeech/"
 
en = utils.get_data(path+'/dataset/en')
es = utils.get_data(path+'/dataset/es')
 
print("Shape of english data:", en.shape)
print("English Data:")
display(en.head(2))
 
print("Shape of spanish data:", es.shape)
print("Spanish Data:")
display(es.head(2))
 
x_en, y_en = en['tweet'].tolist(), en['label'].tolist()
x_es, y_es = es['tweet'].tolist(), es['label'].tolist()
```

```python
from tqdm.notebook import tqdm

def preprocessing(text):
    text = text.replace("<TWEET>", ' \n ')
    return text
 
x_en_cleaned = [preprocessing(X) for X in tqdm(x_en)]
 
x_es_cleaned = [preprocessing(X) for X in tqdm(x_es)]
```

# Cross-Validation

```python
import crossvalidator
```

```python
crossvalidator.cross_validator(x_en_cleaned, y_en, cv=5, ngram_range=(2,3), lang='en')
```

```python
crossvalidator.cross_validator(x_es_cleaned, y_es, cv=5, ngram_range=(3,4), lang='es')
```
